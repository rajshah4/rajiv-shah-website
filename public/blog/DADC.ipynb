Are you looking for better training data for your models? Let me tell
you about dynamic adversarial data collection!

I had a large enterprise customer asking me to incorporate this workflow
into a [Hugging Face](https://www.linkedin.com/company/huggingface/)
private hub demo. Here are some resources I found useful: [Chris
Emezue](https://www.linkedin.com/in/ACoAADC2ZecBGpOHE1kqHIx4NINercY4WG0IkJs)
put together a blog post: “[How to train your model dynamically using
adversarial data](https://huggingface.co/blog/mnist-adversarial)” and a
real-life example using [MNIST using
Spaces](https://huggingface.co/spaces/chrisjay/mnist-adversarial).

If you want an academic paper that details this process, check out:
[Analyzing Dynamic Adversarial Training Data in the
Limit](https://arxiv.org/abs/2110.08514). By using this approach, this
paper found models made 26% fewer errors on the expert-curated test set.
