# SportVu Analysis

### Introduction

This post shares some of the code that I have created for analyzing [NBA
SportVu](http://www.stats.com/sportvu/sportvu-basketball-media/) data.

For background, the [NBA SportVu
data](http://www.stats.com/sportvu/sportvu-basketball-media/) is motion
data for the basketball and players taken 25 times a second. For a
typical NBA game, this means about 2 million rows of data. The data for
over 600 NBA games (first half of the 2015-2016 season) [is
available](https://github.com/rajshah4/BasketballData/). This is over a
billion rows of telematics (iOT) type data. This is a gold mine and here
are some early pieces from studying that data.

### EDA

The first is basic
[EDA](http://projects.rajivshah.com/sportvu/EDA_NBA_SportVu.html) on the
movement data. This code allows you to start analyzing the ball and
player movement.
![basicbb](http://projects.rajivshah.com/images/basicbb.png)

### PBP

The next markdown,
[PBP](http://projects.rajivshah.com/sportvu/PBP_NBA_SportVu.html), shows
how to merge play by play data with the SportVu movement data. This
allows using the annotated data which contains information on the type
of play, score, and home/visitor info.

### Chull

![chull](http://projects.rajivshah.com/images/chull.png) The next set of
documents start analyzing the data. The first measures player spacing
using [convex
hulls](http://projects.rajivshah.com/sportvu/Chull_NBA_SportVu.html).
The next shows how to calculate [player velocity, acceleration, and
jerk](http://projects.rajivshah.com/sportvu/Velocity_NBA_SportVu.html).
(I really wanted to do a post on the biggest jerk in the NBA, but
unfortunately the jerk data is way too noisy.)
![traj](http://projects.rajivshah.com/images/traj.png) The third
document offers a few different ways for analyzing [player and ball
trajectories](http://projects.rajivshah.com/sportvu/Traj_NBA_SportVu.html).

You can find all these files at my [SportVu
Github](https://github.com/rajshah4/NBA_SportVu) repo.
